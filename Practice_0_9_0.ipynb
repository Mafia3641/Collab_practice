# Задание 0

Создайте функцию по нахождению уникальных элементов из двух списков



```
a = [1, 2, 3, 4]
b = [3, 4, 5, 6]
```



````python
a = [1, 2, 3, 4]
b = [3, 4, 5, 6]


def similar(a, b):
	set_a = list(set(a))
	set_b = list(set(b))
	unic_nms = []
	for i in set_a:
		if i not in set_b:
			unic_nms.append(i)
	for j in set_b:
		if j not in set_a and j not in unic_nms:
			unic_nms.append(j)

	return unic_nms


print(similar(a, b))
````

Симулятор виртуального питомца

Цель: создать виртуальный симулятор домашних животных, в котором пользователи смогут заводить питомцев и ухаживать за ними.

Требования:

- Функция для усыновления питомца (имя, тип, возраст).
- Функция для того, чтобы покормить питомца, поиграть с ним или уложить его спать.
- Функция для отображения состояния питомца (голод, радость, энергия).

````python
import random
from random import choice

pet_types = [
	'dragon',
	'cat',
	'dog',
	'parrot'
]

pet_names = [
	'Meow',
	'Mary',
	'Jonny',
	'Alfred'
]

pet_age = range(1, 10)


def create_new():
	type_ = random.choice(pet_types)
	name = random.choice(pet_names)
	age = random.choice(pet_age)
	curr_stats['hp'] = 100
	curr_stats['food'] = 100
	curr_stats['happiness'] = 100
	return name, type_, age


def use_pet(usage):
	if usage.lower() == 'sleep':
		curr_stats['hp'] = 100
		curr_stats['happiness'] -= 10
		curr_stats['food'] -= 10
		print("Your pets energy is full! ")
		print()
		return

	elif usage.lower() == 'play':
		curr_stats['happiness'] = 100
		curr_stats['food'] -= 10
		curr_stats['hp'] -= 10
		print("Your pets happiness is full now! ")
		print()
		return

	elif usage.lower() == 'food':
		curr_stats['food'] = 100
		curr_stats['happiness'] -= 10
		curr_stats['hp'] -= 10
		print("Your pet now is well-fed! ")
		print()
		return

	else:
		print("Wrong move! ")
	return


def check_stats():
	energy = curr_stats['hp']
	food = curr_stats['food']
	happiness = curr_stats['happiness']
	return energy, food, happiness


curr_pet = []
curr_stats = {'hp': 100, 'food': 100, 'happiness': 100}


while True:
    print("Choose your move: ")
    print("1) Create new pet")
    print("2) Give food to your pet")
    print("3) Check pet's stats")
    print()
    move = int(input("Enter a number (1-3) to do a move: "))

    if move == 1:
        new_name, new_type, new_age = create_new()
        curr_pet = [new_name, new_type, new_age]
        print(f"New pet!")
        print(f"Name: {curr_pet[0]}")
        print(f"Type: {curr_pet[1]}")
        print(f"Age: {curr_pet[2]}")
        print()
    elif move == 2:
        if not curr_pet:
            print("You have no pet to use! ")
            print()
        else:
            print("What do you want to do? ")
            print("Let your pet go sleep (sleep)")
            print("Give your pet some food (food)")
            print("Play with your pet (play)")
            print()
            usage = input("Enter your move (from brackets): ")
            print()
            use_pet(usage)
	  elif move == 3:
        energy, food, happiness = check_stats()
        print("Your pet has: ")
        print(f"Energy: {energy}")
        print(f"Food: {food}")
        print(f"Happiness: {happiness}")


````
